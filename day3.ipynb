# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

# line plot

```python
month=['Jan','Feb','Mar','Apr','May']
growth=[4.5,5,5.3,5,6]
```

```python
plt.plot(month,growth)
```

```python
plt.plot(month,growth,'o--')
```

```python
plt.plot(month,growth,'o-')
```

```python
plt.plot(month,growth,'o-',color='red')
plt.xlabel('month')
plt.ylabel('growth in %')
plt.show()  # line aayako hatauxa
```

```python
plt.plot(month,growth,'v-',color='red')
plt.xlabel('month')
plt.ylabel('growth in %')
plt.title("growth by Month")
plt.show() 
```

# Bar Plot

```python
df=sns.load_dataset('iris')
df.head(3)
```

```python
df.species.unique()
```

```python
s=df.species.value_counts()
```

```python
s
```

```python
plt.bar(x=s.index,height=s.values,color=["green",'red','purple'])
plt.xlabel("Species")
plt.ylabel("frequecy")
plt.show()
```

```python
sns.countplot(x='species',data=df)
```

```python
sns.countplot(data=df)
```

# piechart

```python
s=df.species.value_counts()
plt.pie(x=s.values,labels=s.index)
plt.show()
```

```python
s=df.species.value_counts()
plt.pie(x=s.values,labels=s.index,autopct='%.2f%%',colors=['red','pink','purple'])
plt.show()
```
